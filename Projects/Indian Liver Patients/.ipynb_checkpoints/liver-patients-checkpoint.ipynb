```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
warnings.warn("this will not show")
```

```python
df = pd.read_csv("Indian Liver Patient Dataset (ILPD).csv")
df
```

```python
df1 = df.copy()
```

```python
df.info()
```

```python
df.describe().T  # Generates statistical summary and transposes rows into columns.
```

```python
df.isnull().sum()    # Counts the number of missing (NaN) values in each column.
```

```python
df.columns
```

```python
df.columns = df.columns.str.title().str.replace(" ", "_")
df.columns
```

```python
df['Alkphos'].fillna(df['Alkphos'].median(), inplace=True)    # Fills missing values in "Alkphos" with its median value.
```

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Select columns with numerical data (float64 and int64 types)
plt.figure(figsize = (15, 12))

# Loop through each numerical column to create histograms
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(4, 3, i)  # Create a subplot grid (4 rows, 3 columns)
    df[col].hist(bins=20, edgecolor='black')  # Plot histogram with 20 bins
    plt.title(f'Histogram of {col}')  # Set title with column name
    plt.xlabel(col)  # Label x-axis with column name
    plt.ylabel('Frequency')  # Label y-axis as frequency

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```

```python
# Set figure size for better visualization
plt.figure(figsize=(15, 12))

# Loop through each numerical column to create box plots
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(4, 3, i)  # Create a subplot grid (4 rows, 3 columns)
    sns.boxplot(data=df, x=col, color='lightblue')  # Create a box plot for the column
    plt.title(f'Box Plot of {col}')  # Set title with column name
    plt.xlabel(col)  # Label x-axis with column name

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```

```python
# Compute the correlation matrix for numerical columns
correlation_matrix = df[numeric_columns].corr()

# Set figure size for better visualization
plt.figure(figsize=(12, 10))

# Create a heatmap to visualize correlations
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)

# Set the title of the heatmap
plt.title('Correlation Matrix')
plt.show()
```

```python
# Set figure size for better visualization
plt.figure(figsize=(10, 6))

# Create a pie chart of gender counts
df["Gender"].value_counts().plot(kind = "pie", autopct='%.0f%%')
# Set the title of the heatmap
plt.title("Gender Counts")
plt.show()
```

```python
# Set figure size for better visualization
plt.figure(figsize=(8, 4))

# Create a histogram for the 'Age' column
sns.histplot(df['Age'], bins=20, kde=False, color='skyblue', edgecolor='black')

# Set the title of the histogram
plt.title("Age Distribution")

# Label the X-axis as Age
plt.xlabel("Age")

# Label the Y-axis as Frequency
plt.ylabel("Frequency")
plt.show()
```

```python
# Calculate the 95th percentile value for 'Tot_Bilirubin'
tot_bilirubin_threshold = df['Tot_Bilirubin'].quantile(0.95)

# Filter patients with 'Tot_Bilirubin' values greater than the threshold
high_tot_bilirubin_patients = df[df['Tot_Bilirubin'] > tot_bilirubin_threshold]

# Print the number of patients with high 'Tot_Bilirubin' values
print(f"Number of high value patients in 'Tot_Bilirubin': {high_tot_bilirubin_patients.shape[0]}")
```

```python
# Set figure size for better visualization
plt.figure(figsize=(8, 6))

# Create a boxplot for the 'Tot_Bilirubin' column
sns.boxplot(x=df['Tot_Bilirubin'], palette='Set2')

# Add a vertical line for the 95th percentile threshold
plt.axvline(x=tot_bilirubin_threshold, color='red', linestyle='--', label='95th Percentile')

# Set the title of the plot
plt.title('Tot_Bilirubin Distribution with High Value Patients')

# Label the X-axis as 'Tot_Bilirubin'
plt.xlabel('Tot_Bilirubin')

# Display the legend for the red line
plt.legend()
plt.show()
```

```python
# Set figure size for better visualization
plt.figure(figsize=(10, 6))

# Create a scatter plot for high Tot_Bilirubin patients, colored by Gender
sns.scatterplot(data=high_tot_bilirubin_patients, x='Age', y='Tot_Bilirubin', hue='Gender', palette='Set2')

# Set the title of the plot
plt.title('High Tot_Bilirubin Patients: Age vs Tot_Bilirubin (Hue by Gender)')

# Label the X-axis as 'Age'
plt.xlabel('Age')

# Label the Y-axis as 'Tot_Bilirubin'
plt.ylabel('Tot_Bilirubin')

# Display the legend with the title 'Gender'
plt.legend(title='Gender')
plt.show()
```

```python
# Calculate the 95th percentile value for 'Alkphos'
alkphos_threshold = df['Alkphos'].quantile(0.95)

# Filter patients with 'Alkphos' values greater than the threshold
high_alkphos_patients = df[df['Alkphos'] > alkphos_threshold]

# Print the number of patients with high 'Alkphos' values
print(f"Number of high value patients in 'alkphos': {high_alkphos_patients.shape[0]}")
```

```python
# Set figure size for better visualization
plt.figure(figsize=(10, 6))

# Create a scatter plot for high Alkphos patients, colored by Gender
sns.scatterplot(data=high_alkphos_patients, x='Age', y='Alkphos', hue='Gender', palette='Set2')

# Set the title of the plot
plt.title('High Alkphos Patients: Age vs Alkphos (Hue by Gender)')

# Label the X-axis as 'Age'
plt.xlabel('Age')

# Label the Y-axis as 'Alkphos'
plt.ylabel('Alkphos')

# Display the legend with the title 'Gender'
plt.legend(title='Gender')
plt.show()
```

```python
# Set figure size for better visualization
plt.figure(figsize=(8, 6))

# Create a boxplot to show Alkphos distribution by Gender for high Alkphos patients
sns.boxplot(x='Gender', y='Alkphos', data=high_alkphos_patients, palette='Set2')

# Set the title of the plot
plt.title('Alkphos Distribution by Gender for High Alkphos Patients')

# Label the X-axis as 'Gender'
plt.xlabel('Gender')

# Label the Y-axis as Alkphos
plt.ylabel('Alkphos')
plt.show()
```

```python
# Set figure size for better visualization
plt.figure(figsize=(8, 6))

# Create a violin plot to show Alkphos distribution by Gender for high Alkphos patients
sns.violinplot(x='Gender', y='Alkphos', data=high_alkphos_patients, palette='Set2')

# Set the title of the plot
plt.title('Alkphos Distribution by Gender for High Alkphos Patients (Violin Plot)')

# Label the X-axis as 'Gender'
plt.xlabel('Gender')

# Label the Y-axis as 'Alkphos'
plt.ylabel('Alkphos')
plt.show()
```

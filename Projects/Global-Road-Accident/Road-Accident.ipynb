```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

file_path = "road_accident_dataset.csv"
df = pd.read_csv(file_path)

# Display basic information about the dataset
print("Dataset Overview:")
print(df.info())
print("\nFirst 5 Rows:")
print(df.head())

# Summary statistics for numerical columns
print("\nSummary Statistics:")
print(df.describe())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Distribution of categorical features
categorical_columns = [
    'Country', 'Month', 'Day of Week', 'Time of Day', 'Urban/Rural', 
    'Road Type', 'Weather Conditions', 'Driver Age Group', 
    'Driver Gender', 'Vehicle Condition', 'Accident Severity', 'Road Condition', 'Region'
]

for col in categorical_columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette='Set2')
    plt.title(f'Distribution of {col}', fontsize=14)
    plt.xticks(rotation=45)
    plt.xlabel(col, fontsize=12)
    plt.ylabel("Count", fontsize=12)
    plt.tight_layout()
    plt.show()

# Distribution of numerical features
numerical_columns = [
    'Visibility Level', 'Number of Vehicles Involved', 'Speed Limit', 
    'Driver Alcohol Level', 'Pedestrians Involved', 'Cyclists Involved', 
    'Number of Injuries', 'Number of Fatalities', 'Emergency Response Time', 
    'Traffic Volume', 'Medical Cost', 'Economic Loss', 'Population Density'
]

for col in numerical_columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df[col], kde=True, bins=30, color='blue')
    plt.title(f'Distribution of {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
    plt.tight_layout()
    plt.show()

# Accident Severity by Country
plt.figure(figsize=(15, 7))
severity_country = df.groupby(['Country', 'Accident Severity']).size().unstack()
severity_country.plot(kind='bar', stacked=True, colormap='viridis', figsize=(15, 7))
plt.title("Accident Severity by Country", fontsize=16)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.tight_layout()
plt.show()

# Time of Day vs Accident Severity
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Time of Day', hue='Accident Severity', palette='pastel')
plt.title('Accident Severity by Time of Day', fontsize=14)
plt.xlabel('Time of Day', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Severity', fontsize=10)
plt.tight_layout()
plt.show()

# Distribution of fatalities and injuries
plt.figure(figsize=(14, 6))
sns.boxplot(data=df, x='Accident Severity', y='Number of Fatalities', palette='coolwarm')
plt.title('Number of Fatalities by Accident Severity', fontsize=14)
plt.xlabel('Accident Severity', fontsize=12)
plt.ylabel('Number of Fatalities', fontsize=12)
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 6))
sns.boxplot(data=df, x='Accident Severity', y='Number of Injuries', palette='coolwarm')
plt.title('Number of Injuries by Accident Severity', fontsize=14)
plt.xlabel('Accident Severity', fontsize=12)
plt.ylabel('Number of Injuries', fontsize=12)
plt.tight_layout()
plt.show()

# Pedestrians and Cyclists Involved Analysis
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Pedestrians Involved', y='Cyclists Involved', ci=None, palette='magma')
plt.title('Relationship between Pedestrians and Cyclists Involved', fontsize=14)
plt.xlabel('Number of Pedestrians Involved', fontsize=12)
plt.ylabel('Number of Cyclists Involved', fontsize=12)
plt.tight_layout()
plt.show()

# Economic Loss by Region
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Region', y='Economic Loss', ci=None, palette='viridis')
plt.title('Average Economic Loss by Region', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Economic Loss', fontsize=12)
plt.tight_layout()
plt.show()
```

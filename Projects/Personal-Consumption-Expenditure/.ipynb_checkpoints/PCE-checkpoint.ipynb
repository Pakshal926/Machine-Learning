```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "observation_date": [
        "1959-01-01", "1959-02-01", "1959-03-01", "1959-04-01",
        "1959-05-01", "1959-06-01", "1959-07-01", "1959-08-01", "1959-09-01"
    ],
    "PCE": [306.1, 309.6, 312.7, 312.2, 316.1, 318.2, 317.8, 320.2, 324.2],
    "PERCENTAGE": [1092855.00, 1094580.00, 1096135.00, 1097660.00, 1099355.00, 1101010.00, 1102490.00, 1104160.00, 1105910.00]
}

df = pd.DataFrame(data)
df["observation_date"] = pd.to_datetime(df['observation_date'])

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.bar(df["observation_date"], df["PCE"], color="skyblue", label="PCE", alpha=0.7)
ax1.set_xlabel("Observation Date")
ax1.set_ylabel("PCE", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")
ax1.set_xticks(df["observation_date"])
ax1.set_xticklabels(df["observation_date"].dt.strftime("%Y-%m"), rotation=45)

ax2 = ax1.twinx()
ax2.plot(df["observation_date"], df["PERCENTAGE"], color="red", marker="o", label="PERCENTAGE")
ax2.set_ylabel("Percentage", color="red")
ax2.tick_params(axis="y", labelcolor="red")

fig.suptitle("PCE and Percentage Observations Over Time", fontsize=14)
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12, 7))

bar_width = 15
ax1.bar(df["observation_date"], df["PCE"], color="purple", width=bar_width, label="PCE", alpha=0.8, edgecolor="black")
ax1.set_xlabel("Observation Date")
ax1.set_ylabel("PCE", color="purple")
ax1.tick_params(axis="y", labelcolor="purple")
ax1.set_xticks(df["observation_date"])
ax1.set_xticklabels(df["observation_date"].dt.strftime("%y-%m"), rotation=45)

ax2 = ax1.twinx()
ax2.plot(df["observation_date"], df["PERCENTAGE"], color="green", marker="o", label="PERCENTAGE", linestyle="--", linewidth=2)
ax2.set_ylabel("Percentage", color="green")
ax2.tick_params(axis="y", labelcolor="green")

fig.suptitle("PCE and Percentage Observation Over Time", fontsize=14)
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.tight_layout()
plt.show()
```

```python
import numpy as np

categories = df["observation_date"].dt.strftime("%Y-%m").tolist()
values = df["PERCENTAGE"].tolist()

num_vars = len(categories)

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

values += values[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw = dict(polar=True))
ax.fill(angles, values, color="teal", alpha=0.25)
ax.plot(angles, values, color="teal", linewidth=2)

ax.set_yticks([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(values[:-1], fontsize=9, color="darkgreen")

ax.set_title("Percentage Observations in Radar Chart", fontsize=14, pad=20)

plt.show()
```
